```python
import os
from collections import OrderedDict
import pickle
```

# Loading Data

```python
# reading the data from the file
with open('corpus_index_posting.txt', 'rb') as handle:
    data = handle.read()
  
print("Data type before reconstruction : ", type(data))
  
# reconstructing the data as dictionary
corpus = pickle.loads(data)
  
print("Data type after reconstruction : ", type(corpus))
print(len(corpus))
```

# Reading Test.txt

```python
# open the file in read mode
with open('Test.txt', 'r', encoding='utf-8') as file:
    # read all the words in the file and split them by whitespace
    terms = file.read()
    terms = terms.replace('\ufeff', '')
    terms = terms.split()
```

```python
for term in terms:
    if term in corpus:
        print(str(term),': ')
        print('Document frequency',corpus[term][0])
        for k in corpus[term][1]:
            print(k)
        print()
    else:
        print('Term not Found\n')
```
